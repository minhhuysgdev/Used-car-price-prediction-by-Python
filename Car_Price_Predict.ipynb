## 4. Data Preparation (Chuẩn hóa dữ liệu để phù hợp): Những trường dữ liệu gì sẽ hữu ích với mục tiêu của chúng ta? Tại sao? Loại bỏ những trường, dòng dữ liệu không hữu ích. Tạo ra cột dữ liệu mới bla bla

```python
car_data.info()
```

```python
car_data.isnull().sum() #kiem tra du lieu NULL
```

```python
car_data
```

```python
 #manual encoding 
car_data.replace({'Fuel_Type':{'Petrol':0,'Diesel':1,'CNG':2}},inplace=True)
#one hot encoding 
car_data=pd.get_dummies(car_data, columns=['Seller_Type','Transmission'],drop_first=True)
```

```python
car_data
```

```python
X=car_data.drop(['Car_Name','Selling_Price'],axis=1)
y=car_data['Selling_Price']
```

```python
X
```

```python
print("Shape of X is",X.shape)
print("Shape of y is",y.shape)
```

```python
X_train, X_test, y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
print(" X_test shape is",X_test.shape)
print(" X_train shape is",X_train.shape)
print(" y_test shape is",y_test.shape)
print(" y_test shape is",y_train.shape)
```

```python
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
X_test=scaler.fit_transform(X_test)
```

```python
X_train
```

```python
X_test
```

## 5. Data Modeling: Sử dụng các thuật toán khác nhau: Kmean, Decision tree, … với các perspective khác nhau để tìm ra các model/pattern khác nhau. Trình bày các thuật toán được sử dụng

## Linear Regression Model

```python
#Construct object 
from sklearn.linear_model import LinearRegression
lnr_model=LinearRegression()
```

```python
#Training model 
lnr_model.fit(X_train,y_train)
```

```python
lnr_pred=lnr_model.predict(X_test)
```

## Decision Tree Regression Model

```python
from sklearn.tree import DecisionTreeRegressor
dt_model=DecisionTreeRegressor()
```

```python
#fit traning data into model
dt_model.fit(X_train,y_train)
```

```python
dt_pred=dt_model.predict(X_test)
```

## Random Forest Regression Model

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf_model=RandomForestRegressor()
```

```python
rf_model.fit(X_train,y_train)
```

```python
rf_pred=rf_model.predict(X_test)
```

## 6. Data Evaluation: Đánh giá kết quả đối với các thuật toán cần training. Còn các thuật toán tuyến tính (Kmean,…) chỉ cần visualization dữ liệu lên là được.


```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score 
```

```python
sns.regplot(x=lnr_pred,y=y_test,scatter_kws={"color":"blue"}, line_kws={"color":"red"})
plt.xlabel("Predicted Price")
plt.ylabel("Actual Price")
plt.title('Linear Regression Prediction Result')
plt.show()
```

```python
#evalute linearRegressor
mae_lnr=metrics.mean_absolute_error(lnr_pred,y_test)
mse_lnr=metrics.mean_squared_error(lnr_pred,y_test)
R2_lnr=metrics.r2_score(lnr_pred,y_test)
print("MAE: ",mae_lnr)
print("MSE: ",mse_lnr)
print("R2 score: ",R2_lnr)
```

## Deployment

```python
import tkinter as tk
from tkinter import *
from  tkinter import ttk
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
```

```python
# tkinter GUI
root= tk.Tk()
root.title('Used Car Price Prediction')
root.configure(bg='lightgrey')
canvas1 = tk.Canvas(root, width = 500, height = 400)
canvas1.pack()
# Fuel Type

label1 = tk.Label(root, text='Fuel Type:')
canvas1.create_window(100, 100, window=label1)

entry1 = tk.Entry(root)
canvas1.create_window(270, 100, window=entry1)

# Transmission Type
label2 = tk.Label(root, text='Transmission Type:         ')
canvas1.create_window(120, 120, window=label2)

entry2 = tk.Entry (root)
canvas1.create_window(270, 120, window=entry2)

# Kms Driven
label3 = tk.Label(root, text='Kms Driven:                       ')
canvas1.create_window(140, 140, window=label3)

entry3 = tk.Entry (root)
canvas1.create_window(270, 140, window=entry3)

# Owner
label4 = tk.Label(root, text='Owner:                                  ')
canvas1.create_window(160, 160, window=label4)

entry4 = tk.Entry (root)
canvas1.create_window(270, 160, window=entry4)

# Year
label5 = tk.Label(root, text='Year:                                         ')
canvas1.create_window(180, 180, window=label5) 

entry5 = tk.Entry (root)
canvas1.create_window(270, 180, window=entry5)


#Seller Type
label6 = tk.Label(root, text='Seller Type:                                        ')
canvas1.create_window(180, 200, window=label6) 
entry6 = tk.Entry (root)
canvas1.create_window(270, 200, window=entry6)

#Seller Type
label7 = tk.Label(root, text='Present Price:                                          ')
canvas1.create_window(180, 220, window=label7) 
entry7 = tk.Entry (root)
canvas1.create_window(270, 220, window=entry7)

def predicts(): 
    global fuel
    fuel = float(entry1.get()) 
    
    global transmission
    transmission= float(entry2.get()) 
    
    global kms
    kms = float(entry3.get()) 
    
    global owner
    owner = float(entry4.get()) 
    
    global year
    year = float(entry5.get()) 
    
    global seller
    seller = float(entry6.get())
    
    global present
    present = float(entry7.get())
    
    ws = Tk()
    ws.title('Predicted Result')
    ws.geometry('400x200')
    set = ttk.Treeview(ws)
    set.pack()

    set['columns']= ('Model_Name','price')
    set.column("#0", width=0,  stretch=NO)
    set.column("Model_Name",anchor=CENTER, width=200)
    set.column("price",anchor=CENTER, width=200)

    set.heading("#0",text="",anchor=CENTER)
    set.heading("Model_Name",text="Model",anchor=CENTER)
    set.heading("price",text="Predicted Price",anchor=CENTER)

    set.insert(parent='',index='end',iid=0,text='', values=('Linear Regression',lnr_model.predict([[fuel,transmission,kms,owner,year,seller,present]])))
    set.insert(parent='',index='end',iid=1,text='', values=('Decision Tree',dt_model.predict([[fuel,transmission,kms,owner,year,seller,present]])))
    set.insert(parent='',index='end',iid=2,text='', values=('Random Forest', rf_model.predict([[fuel,transmission,kms,owner,year,seller,present]])))

button1 = tk.Button(root, text='Used Car Price Prediction',command=predicts, bg='blue', fg='white')
canvas1.create_window(270, 260, window=button1)


root.mainloop()
```
